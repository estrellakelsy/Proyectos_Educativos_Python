```python
import requests
from bs4 import BeautifulSoup
page = requests.get("https://www.bolsamadrid.es/esp/aspx/Portada/Portada.aspx")
soup = BeautifulSoup(page.text, 'html.parser')
table = soup.find_all("table")
for table in table:
    agenda = table.find(align="center")
    print ([agenda, table])
```

```python
import requests
from bs4 import BeautifulSoup
page = requests.get("https://www.bolsamadrid.es/esp/aspx/Portada/Portada.aspx")
soup = BeautifulSoup(page.text, 'html.parser')
table = soup.find_all(id="TblIndPort")
for table in table:
    agenda = table.find(class_="On mclick").text
    agendasemanal = table.find(class_="Off mclick").text 

    print ([agenda,agendasemanal])
```

```python
import requests
from bs4 import BeautifulSoup
page = requests.get("https://www.bolsamadrid.es/esp/aspx/Portada/Portada.aspx")
soup = BeautifulSoup(page.text, 'html.parser')
cuerpo = soup.find('table', tableclass_="TblPort TblIndPort")  
for tabla2 in cuerpo.find_all('Indicadores'):
    fila = tabla2.find_all('tr')
    for filas in fila:
        evento = filas.find('th', class_='TblPort TblIndPort')
        print(evento)
```
