**1. Generar un DataFrame con los datos del fichero.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
data = pd.read_csv('Titanic.csv', sep=',')
data
```

**2. Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas, los tipos de datos de las columnas, las 10 primeras filas y las 10 últimas filas**

```python
# mostrar la informacion  actual de una matriz
print('Dimensiones: ', data.shape)
print('*************************************************--------**\n'*1)
#print('Dimen: ', data.ndim) # muestra 2 dimensiones
print('Numero de elementos:\n', data.size)
print('*************************************************-------**\n'*1)
print('Nombre de las columnas:\n', data.columns)
print('*************************************************-------**\n'*1)
print('Nombre de las filas:\n', data.index)
print('*************************************************-------**\n'*1)
print('Tipos de datos:\n', data.dtypes)
print('*************************************************-------**\n'*1)
print('Primeros 10 filas:\n', data.head(10))
print('*************************************************-------**\n'*1)
print('Ultimas 10 filas:\n', data.tail(10))
```

**4. Mostrar por pantalla las filas pares del DataFrame.**

```python
# iloc se utiliza en los DataFrames para seleccionar los elementos en base a su ubicación
print(data.iloc[range(0,data.shape[0],2)])
```

**5. Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.**

```python
# Pclass es el nombre de la columna
# == 1 indica la clase del pasajero
# Name indica la columna
print(data[data["Pclass"]==1]["Name"].sort_values())
```

**6. Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.**

```python
# de la data columna survived 
# value_counts: devuelve un objeto que contiene recuentos de valores únicos
print(data['Survived'].value_counts()/data['Survived'].count()*100)

# 
print('*************************************************-------**\n'*1)
print(data['Survived'].value_counts(normalize=True)*100)
```

**7. Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.**

```python
# con groupby agrupamos 2 columnas para contar la cantidad de sobrevivientes 
# por clase y con value_counts contamos lo valores para luevo * 100
print(data.groupby('Pclass')['Survived'].value_counts(normalize=True)*100)
```

**9. Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.**

```python
# agrupar la data con groupby, 
# columnas(pclass,sex), 
# vunstack devuelve por orden gerarquico
print(data.groupby(['Pclass','Sex'])['Age'].mean().unstack()['female'])
```
