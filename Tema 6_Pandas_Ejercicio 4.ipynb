**Los ficheros emisiones-2016.csv, emisiones-2017.csv, emisiones-2018.csv y emisiones-2019.csv, contienen datos sobre las emisiones contaminates en la ciudad de Madrid en los a√±os 2016, 2017, 2018 y 2019 respectivamente. Escribir un programa con los siguientes requisitos:**

**1. Generar un DataFrame con los datos de los cuatro ficheros.**


```python
import pandas as pd 
import matplotlib 
import datetime as dt
emisiones2016 = pd.read_csv("emisiones-2016.csv", sep=';')
emisiones2017 = pd.read_csv("emisiones-2017.csv", sep=';')
emisiones2018 = pd.read_csv("emisiones-2018.csv", sep=';')
emisiones2019 = pd.read_csv("emisiones-2019.csv", sep=';')
Datos = pd.concat([emisiones2016,emisiones2017,emisiones2018,emisiones2019])
Datos
```

**2. Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, A√ëO, MES y las correspondientes a los d√≠as D01, D02, etc.**

```python
Columnas = ['ESTACION','MAGNITUD','ANO','MES'] # mostrar las 4 columnas

#startswith: El m√©todo devuelve True si la cadena comienza con el valor especificado; de lo contrario, False.
Columnas.extend([columns  for columns in Datos if columns.startswith('D')])
Datos = Datos[Columnas]
Datos
```

**3. Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los d√≠as aparezcan en una √∫nica columna.**

```python
#melt:  desenvuelve un DataFrame de formato ancho a formato
Datos = Datos.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR')
Datos
```

**4. A√±adir una columna con la fecha a partir de la concatenaci√≥n del a√±o, el mes y el d√≠a (usar el m√≥dulo datetime).**

```python
# paso 1 eliminar el caracter D del comiendo de la columna de los dias
Datos['DIA'] = Datos.DIA.str.strip('D')

# paso 2: Concatenamos las columnas del a√±o, mes y dia
Datos['FECHA'] = Datos.ANO.apply(str) + '/' + Datos.MES.apply(str) + '/' + Datos.DIA.apply(str)

#paso 3: convertir la nueva columna al tipo de fecha
Datos['FECHA'] = pd.to_datetime(Datos.FECHA, format='%Y/%m/%d', infer_datetime_format=True, errors='coerce')
Datos
```
